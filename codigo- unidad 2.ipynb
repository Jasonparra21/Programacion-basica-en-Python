```python
### ESCRIBA SU CÓDIGO A PARTIR DE AQUÍ ### (~ 7-10 líneas de código)

# Entrada del programa (~ 1 línea).
n = int(input())                      

# Serie de Leibniz (~ 5-8 líneas). 
pi = 0      
for x in range(0,n):
  if x % 2 == 0:
    continue
  pi=-pi+(1/x)
pi=abs(4*pi)

    
# Salida del programa con 10 decimales (~ 1 línea).
print(f"{pi:.10f}")
```
